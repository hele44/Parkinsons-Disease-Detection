## Feature Selection


```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
X=df.drop(labels=['status'],axis=1)
Y=df['status']
X.head()
```

```python
feature_sel_model=SelectFromModel(Lasso(alpha=0.0005))
```

```python
feature_sel_model.fit(X,Y)
```

```python
cols=X.columns
```

```python
cols
```

```python
selected_feature=cols[feature_sel_model.get_support()]
```

```python
selected_feature
```

```python
X=X[selected_feature]
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=40)
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(scaled_X_train,Y_train)
```

```python
coefs = pd.Series(index=X.columns,data=log_reg.coef_[0])
```

```python
coefs = coefs.sort_values()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=coefs.index,y=coefs.values);
plt.xticks(rotation=90);
```

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix,accuracy_score,ConfusionMatrixDisplay
from sklearn import metrics
```

```python
test_preds = log_reg.predict(scaled_X_test)
```

```python
cm=confusion_matrix(Y_test, test_preds)
```

```python
disp=ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
print(classification_report(Y_test,test_preds))
```

```python
print("Model accuracy on test is: ", accuracy_score(Y_test, test_preds))
```

```python
def report_model(model):
    model.fit(scaled_X_train,Y_train)
    model_preds = model.predict(scaled_X_test)
    print("confusion_matrix test is: ", confusion_matrix(Y_test, model_preds))
    print(classification_report(Y_test,model_preds))
    print('\n')
    print("Model accuracy on test is: ", accuracy_score(Y_test, model_preds))
    print('\n')
```

## KNN GridSearch

```python
from sklearn.pipeline import Pipeline
```

```python
scaler = StandardScaler()
```

```python
knn = KNeighborsClassifier()
```

```python
operations = [('scaler',scaler),('knn',knn)]
```

```python
pipe = Pipeline(operations)
```

```python
k_values = list(range(1,30))
```

```python
param_grid = {'knn__n_neighbors': k_values}
```

```python
full_cv_classifier = GridSearchCV(pipe,param_grid,scoring='accuracy')
```

```python
full_cv_classifier.fit(X_train,Y_train)
```

```python
model_preds = full_cv_classifier.predict(X_test)
print("confusion_matrix test is: ", confusion_matrix(Y_test, model_preds))
print(classification_report(Y_test,model_preds))
print('\n')
print("Model accuracy on test is: ", accuracy_score(Y_test, model_preds))
print('\n')
```

```python
full_cv_classifier.best_estimator_.get_params()
```

```python
scores = full_cv_classifier.cv_results_['mean_test_score']
plt.plot(k_values,scores,'o-')
plt.xlabel("K")
plt.ylabel("Accuracy");
```
